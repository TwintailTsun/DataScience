## Часть 3: Обработка DataFrame
### Выбор данных
Pandas позволяет выбирать данные по индексам, меткам столбцов и условиям.

```python
# Выбор столбца по метке
# Используйте синтаксис DataFrame['название_столбца']
df_new['Pclass']

# Выбор нескольких столбцов
# Используйте синтаксис DataFrame[['столбец_1', 'столбец_2']]
df_new[['Pclass', 'Age']]

# Выбор строк по индексу
# Используйте метод .loc[]
df_new.loc[2]

# Выбор строк и столбцов по условию
# Используя логические операции, выберите мужчин старше 30
df_new.query("Age > 30 & Sex == 'male'")
```

### Сортировка данных
Сортировка данных по значениям столбцов.

```python
# Сортировка данных по столбцу 'PassengerId' по возрастанию
# Используйте метод .sort_values()
df_new.sort_values(by=['PassengerId'])
```

### Группировка данных
Pandas также позволяет группировать данные и выполнять агрегирующие операции.

```python
# Найдите долю выживших среди всех PClass
# Используйте метод .groupby() для группировки данных по столбцу 'Pclass'
# Затем вычисляем среднее значение столбца 'Survived' для каждой группы
df_new.groupby(by=['Pclass'])['Survived'].mean()
```

## Часть 4: Задания для практики
Прочитайте данные из файла 'titanic.csv'.
Проверьте, есть ли пропущенные значения в данных и заполните их нулями.
Выведите первые 10 строк данных.
Выберите только те строки, где значение в столбце 'Age' больше 30.
Отсортируйте данные по столбцу 'Fare' в порядке убывания.
Сгруппируйте данные по столбцу 'Pclass' и вычислите средний возраст ('Age') для каждого класса.

```python
import pandas as pd

# Используем метод read_csv для чтения данных из файла 'titanic.csv'
data_frame = pd.read_csv('titanic.csv')

# Используем метод isna для проверки наличия пропущенных значений в данных
missing_values = data_frame.isna()
print(missing_values)

# Используем метод fillna для заполнения пропущенных значений нулями
# Параметр inplace=True позволяет изменить исходный DataFrame
data_frame.fillna(0, inplace=True)

# Используем метод head для вывода первых 10 строк данных
print(data_frame.head(10))

# Используем метод query для фильтрации строк, где значение в столбце 'Age' больше 30
filtered_data = data_frame.query('Age > 30')

# Используем метод sort_values для сортировки данных по столбцу 'Fare' в порядке убывания
sorted_data = filtered_data.sort_values(by='Fare', ascending=False)

# Используем метод groupby для группировки данных по столбцу 'Pclass'
# Затем используем метод mean для вычисления среднего возраста ('Age') для каждого класса
average_age_by_class = sorted_data.groupby('Pclass')['Age'].mean()

# Выводим результат
print(average_age_by_class)
```
